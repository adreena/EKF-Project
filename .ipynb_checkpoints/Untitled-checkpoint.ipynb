```python
import matplotlib.pyplot as plt
import csv
```

```python
def draw_data(file_name, title):
    time= []
    px = []
    py = []
    vx = []
    vy = []
    with open(file_name,'r') as csvfile:
    #     plots = csv.reader(csvfile, delimiter=',')
        lis=[line.split() for line in csvfile]
        count = 0
        for line in lis:
            
            row = line[0].split(',')
            if len(row) == 4 and row[-1]!='':
                time.append(count)
                px.append(round(float(row[0]), 2))
                py.append(round(float(row[1]), 2))
                vx.append(round(float(row[2]), 2))
                vy.append(round(float(row[3]), 2))
                count+=1
    plt.plot(time,px,'r',label='px') # plotting t,a separately 
    plt.plot(time,py,'y',label='py') # plotting t,b separately 
    plt.plot(time,vx,'g',label='vx') # plotting t,c separately 
    plt.plot(time,vy,color='#3FAABF',label='vy') # plotting t,c separately 
    plt.xlabel('time')
    plt.ylabel('RMSE')
    plt.legend()
    plt.title(title)
    plt.show()
    plt.close()
```

```python
draw_data('result_1_laser.csv', 'DATA 1 Laser')
draw_data('result_1_radar.csv', 'DATA 1 Radar')
draw_data('result_1.csv', 'DATA 1 Both')
draw_data('result_2_laser.csv', 'DATA 2 Laser')
draw_data('result_2_radar.csv', 'DATA 2 Radar')
draw_data('result_2.csv', 'DATA 2 Both')
```
